# Prey-predator model

## We aim to sole the following equations:

\begin{align}
\frac{dy_1}{dt} &= y_1(a - b y_2) \\
\frac{dy_1}{dt} &= - y_2(c - d y_1)
\end{align}


```python
import numpy as np
import scipy.integrate as sp
import matplotlib.pyplot as plt
```

```python
y = np.array([0,0])
a , b , c , d = 0.25, 0.1, 1, 0.1

def euler_step(y,dt):
    dy1dt = y[0] * (a - b * y[1])
    dy2dt = -y[1] * (c - d * y[0])
    y_ret = y + dt * np.array([dy1dt, dy2dt])
    return y_ret

def run_until_T( y_0, T, N_steps=10000 ):
    dt = T / N_steps
    y = np.zeros((N_steps+1, 2))
    y[0] = y_0
    for step in range(1,N_steps+1):
        y[step] = euler_step(y[step-1], dt)
    return y

y_0 = np.array([5, 5])
T = 100
y = run_until_T(y_0, T)
plt.figure(figsize=(10,5))
plt.plot(y[:,0], label='prey')
plt.plot(y[:,1], label='predator')
plt.legend()
plt.xlabel('time')
plt.ylabel('population')
plt.title('Prey-Predator model')
plt.show()
```
